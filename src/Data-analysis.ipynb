```python
from collections import Counter
```

```python
fr = dict(Counter(UID))
```

```python
plt.figure(figsize=(12,6))
sns.distplot(fr.values())
plt.title('Distribution of number of ratings from users',fontdict={'fontsize':30})
plt.show()
```

From the plot we can understand that there are definitely some users who rated alot of movies

```python
max(fr.values())
```

So there is a user who has rated 18276 movies. We would try to remove such user and then again try to plot the above graph

```python
fre = sorted(fr.values())
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(fre,cumulative=True)
plt.title('CDF of number of ratings from users',fontdict={'fontsize':30})
plt.show()
```

```python
fre = fre[:int(0.99*len(fre))]
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(fre,cumulative=True)
plt.title('CDF of number of ratings from users',fontdict={'fontsize':30})
plt.show()
```

The users who had crazy amout of ratings have been removed, now we will plot the distribution of number of ratings by the users

```python
plt.figure(figsize=(12,6))
sns.distplot(fre)
plt.title('Distribution of number of ratings from users',fontdict={'fontsize':30})
plt.show()
plt.figure(figsize=(12,6))
sns.distplot(fre)
plt.title('Distribution of number of ratings from users',fontdict={'fontsize':30})
plt.xlim(0,300)
plt.show()
```

I zoomed in on the range (0,300) to check how the distribution looks.<br>
So approximately 0-50 has the maximum percentage in the histogram, so most users have rated 1-50 times.

The MovieIDs are not continious and therfore we would make them continuous. We would also make the userId continuous

```python
df.columns
```

```python
import scipy as sp
df = df.sort_values(by='userId')
CID = sp.array(df.userId)
count = 0; flag = CID[0]
for i in range(CID.shape[0]):
    if(CID[i]==flag):
        CID[i]=count
    else:
        count += 1
        flag = CID[i]
        CID[i]=count
df['userId'] = CID
df = df.sort_values(by='movieId')
MID = sp.array(df.movieId)
count = 0; flag = MID[0]
for i in range(MID.shape[0]):
    if(MID[i]==flag):
        MID[i]=count
    else:
        count += 1
        flag = MID[i]
        MID[i]=count
df['movieId'] = MID
```
